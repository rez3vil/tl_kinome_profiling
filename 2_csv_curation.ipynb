```python
import pandas as pd
import os
```

```python
# Load the CSV file
input_file = 'dataset_train_target_191_v2.csv'  # Replace with your actual file name
df = pd.read_csv(input_file)
```

```python
# Get the list of label columns (assuming the last 10 columns are the labels)
label_columns = df.columns[-191:]
feature_columns = df.columns[:-191]
```

```python
# Create a directory to store the individual files
output_dir = 'split_csv_files'
os.makedirs(output_dir, exist_ok=True)
```

```python
# Step 1: Split the CSV file into 191 files, each containing the same features and 1 label
for i, label in enumerate(label_columns):
    split_df = df[feature_columns.tolist() + [label]]
    split_df = split_df.dropna(subset=[label])
    split_df["Kinase_family"] = label
    split_df.rename(columns={label: 'Activity'}, inplace=True)
    output_file = os.path.join(output_dir, f'file_with_label_{i+1}.csv')
    split_df.to_csv(output_file, index=False)
    print(f'Generated {output_file}')
```

```python
concatenated_df = pd.DataFrame()

for i, label in enumerate(label_columns):
    # Load each individual CSV
    file_path = os.path.join(output_dir, f'file_with_label_{i+1}.csv')
    temp_df = pd.read_csv(file_path)

    # Concatenate to the master dataframe
    concatenated_df = pd.concat([concatenated_df, temp_df])
```

```python
# Optional: Save the concatenated DataFrame into a final CSV file
concatenated_file = 'dataset_train_target_191_v3.csv'
concatenated_df.to_csv(concatenated_file, index=False)
print(f'Concatenated file saved as {concatenated_file}')
```
